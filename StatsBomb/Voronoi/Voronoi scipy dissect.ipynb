# Dissect the Scipy Voronoi diagram

```python
import numpy as np

from scipy.spatial import Voronoi, voronoi_plot_2d
```

### Randomly generated Voronoi points

```python
# Create random points
#points = np.random.randint(3, size=(10,2))

points = np.array([[0, 0], [0, 1], [0, 2], [1, 0], [1, 1], [1, 2],

                   [2, 0], [2, 1], [2, 2]])
points
```

```python
# Set the points into a Voronoi Object
vor = Voronoi(points)
vor
```

### Plotting the Voronoi points

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = voronoi_plot_2d(vor)

```
