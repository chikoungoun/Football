```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('df_api_all.csv',index_col=0)
```

```python
df
```

```python
m1 = df.loc[df['match_id'] == 3853754]
```

```python
m1
```

```python
m1['home_team'] = m1.apply(lambda row : 1 if row['team.name'] =='Raja Casablanca'  else 0, axis=1)
```

```python
m1
```

```python
m1.columns.to_list()
```

```python
shots = m1.loc[m1['type.name'] == 'Shot']
```

```python
shots
```

```python
home_shots = shots.loc[shots['home_team'] == 1,['minute','second','xg']]
home_shots['time'] = home_shots['minute'].astype(str)+":"+home_shots['second'].astype(str)
home_shots['xg_cum'] = round(home_shots['xg'].cumsum(),3)
home_shots
```

```python
away_shots = shots.loc[shots['home_team'] == 0,['minute','second','xg']]
away_shots['time'] = away_shots['minute'].astype(str)+":"+away_shots['second'].astype(str)
away_shots['xg_cum'] = round(away_shots['xg'].cumsum(),3)
away_shots
```

### Plot

* Scaling the display of the y values

```python
y_tick_max = max(home_shots['xg_cum'].max(),away_shots['xg_cum'].max())
y_tick_max
```

```python
ticks = np.linspace(0, y_tick_max, num=8)
ticks
```

```python
fig, ax = plt.subplots(figsize = (16,8))


plt.ylim(0,round(y_tick_max))

plt.xticks([0,15,30,45,60,75,90])
plt.yticks(ticks)
plt.vlines( ymin=0, ymax=4,x=45, color='red', alpha=0.5,linestyle="dotted")


plt.xlabel('minute')
plt.ylabel('xG')
plt.grid()

ax.step(x=home_shots['minute'],y=home_shots['xg_cum'],marker=".",where='post')
ax.step(x=away_shots['minute'],y=away_shots['xg_cum'],marker=".",where='post')


plt.show()
```
