# ************************** DATA PREPROCESSING ********************************

## A. Data Preprocessing

### A.1 Add pass, dribble, carries, pressure, interception, foul commited and shot columns

```python
dummy = pd.get_dummies(df['type.name'])


#Control if some events did not happen
one_hot = ['Pass','Dribble','Carry','Pressure','Block','Interception','Foul Committed','Shot','Own Goal Against']

for i in one_hot:
    if i not in dummy.columns.to_list():
        one_hot.remove(i)

dummy = pd.get_dummies(df['type.name'])[one_hot]
dummy_name = []
for i in dummy.columns.tolist()[-10:]:
    dummy_name.append(i.lower())
dummy.columns=dummy_name
#ACHTUNG
#df=df.join(dummy)
df = pd.concat([df,dummy],axis=1)
```

### A.2 add goal column

```python
df['goal'] = df.apply(lambda row : 1 if row['type.name']=='Goal' or (row['type.name'] == 'Own Goal For') else 0,axis=1)
```

### A.3 add xg column

```python
df['xg'] = df.apply(lambda row : row['shot.statsbomb_xg'] if (row['type.name'] == 'Shot') else 0,axis=1)
```

## B. Defensive actions

### B.1. Duels

#### B.1.1 duel column

```python
df['duel']=df.apply(lambda row:1 if 
                        ((row['clearance.aerial_won'] == True) or (row['pass.aerial_won']==1) or (row['clearance.aerial_won']==True)) 
                        or 
                        (row['type.name']=='Duel') 
                        else 0,axis=1)
```

#### B.1.2 duel_won column

```python
duel_win_criteria = ['Success In Play','Success Out','Lost Out','Won']
df['duel_won'] = df.apply(lambda row: 1 if 
                          (row['type.name'] == 'Duel' and row['duel.outcome.name'] in duel_win_criteria)
                          or
                          ((row['clearance.aerial_won'] == True) or (row['pass.aerial_won']==1) or (row['clearance.aerial_won']==True))
                          else 0,axis=1)
```

### B.2 Aerial Duels

#### B.2.1 aerial_duel column

```python
df['aerial_duel']=df.apply(lambda row:1 if 
                        ((row['clearance.aerial_won'] == True) or (row['pass.aerial_won']==1) or (row['clearance.aerial_won']==True)) 
                        or 
                        (row['type.name']=='Duel'and row['duel.outcome.name']=='Aerial Lost') 
                        else 0,axis=1)
```

#### B.2.2 aerial_duel_won column¶

```python
df['aerial_duel_won']=df.apply(lambda row:1 if 
                        ((row['clearance.aerial_won'] == True) or (row['pass.aerial_won']==1) or (row['clearance.aerial_won']==True))
                        else 0,axis=1)
```

#### B.2.3 aerial_duel_lost column

```python
df['aerial_duel_lost']=df.apply(lambda row:1 if 
                        (row['type.name']=='Duel'and row['duel.outcome.name']=='Aerial Lost') 
                        else 0,axis=1)
```

### B.3 Interceptions

#### B.3.1 interception_won column

```python
df['interception_won']=df.apply(lambda row:1 if 
                        (row['interception.outcome.name'] == True) else 0,axis=1)
```

#### B.3.2 interception_lost column

```python
df['interception_lost']=df['interception']-df['interception_won']
```

## C. Progresive Actions

### C.1. Passes

#### C.1.1 pass_acc column

```python
#Replacing outcome_name's empty cells by value 1 for Pass events ONLY! (means complete passes)
df.loc[(pd.isna(df['pass.outcome.name']) == True) & (df['pass'] == 1) ,'pass.outcome.name'] = 1

#Creating Pass_Acc Column
df['pass_acc']=df['pass']*df['pass.outcome.name']

#Replacing values different than 1 by 0
df.loc[(df['pass_acc'] != 1) ,'pass_acc']= 0
```

#### C.1.2 pass_not_acc column

```python
df['pass_not_acc']=df['pass']-df['pass_acc']
```

### C.2 Progressive actions

#### C.2.1 Distances calculation

```python
#create a concealed pass & carry end_location
#start with the pass end_location
df['end_location'] = df['pass.end_location']
```

```python
#add carry.end_location
df['end_location'].fillna(df['carry.end_location'],inplace=True)
```

* recreate the location_x/location_y and end_location_x/end_location_y columns

```python
# location_x & location_y
df['location_x'] = df['location'].str[0]
df['location_y'] = df['location'].str[1]

# end_location_x & end_location_y
df['end_location_x'] = df['end_location'].str[0]
df['end_location_y'] = df['end_location'].str[1]
```

```python
#Creating Dist_Beginning and Dist_End Columns
df['dist_beginning'] = np.sqrt(np.square(120-df['location'].str[0]) + np.square(40 - df['location'].str[1]))
df['dist_end']=np.sqrt(np.square(120 - df['end_location'].str[0]) + np.square(40 - df['end_location'].str[1]))
```

#### C.2.2 Prog calculation

```python
df['prog']=0
df.loc[ df['dist_end']<0.75*df['dist_beginning'] ,'prog']= 1
```

### C.3 Progressive Passes

#### C.3.1 pass_prog column

```python
df['pass_prog']=df['pass']*df['prog']
```

#### C.3.2 pass_prog_acc column

```python
df['pass_prog_acc']=df['pass_prog']*df['pass.outcome.name']
#Replacing values different than 1 by 0
df.loc[ (df['pass_prog_acc'] != 1) ,'pass_prog_acc']= 0
```

#### C.3.3 pass_prog_not_acc

```python
df['pass_prog_not_acc']=df['pass_prog']-df['pass_prog_acc']
```

### C.4 Progressive Carries

#### C.4.1 carries_prog

```python
df['carry_prog'] = df['carry']*df['prog']
```

#### C.4.2 carries_not_prog

```python
df['carry_not_prog']=df['carry']-df['carry_prog']
```

### C.5 Dribbles

##### C.5.1 dribble_acc column

```python
#Replacing outcome_name's equql to 'Complete'  by value 1 for Dribble events ONLY!
df.loc[(df['dribble.outcome.name'] == 'Complete') & (df['dribble'] == 1) ,'dribble.outcome.name']= 1

#Creating Dribble_Acc Column
df['dribble_acc']=df['dribble']*df['dribble.outcome.name']

#Replacing NAN values in Column 'Dribble_ACC' by 0 value
df['dribble_acc'] = df['dribble_acc'].replace(np.nan, 0)
```

## D. Change_Possession

```python
def change_possession_event(df):
    df['change_possession']=0
    for i in range(0,len(df['type.name'])-1):
        if df['possession_team.name'].iloc[i]!=df['possession_team.name'].iloc[i] and df['possession_team.name'].iloc[i+1]!=df['possession_team.name'].iloc[i]:
            df['change_possession'].iloc[i]=1
    return df
```

```python
# Spliting the x locations into different zones
df['zones']=0
df.loc[ (df['location'].str[0]<=20) ,'zones']= 1
df.loc[ (df['location'].str[0]>20) & (df['location'].str[0]<=40),'zones']= 2
df.loc[ (df['location'].str[0]>40) & (df['location'].str[0]<=60),'zones']= 3
df.loc[ (df['location'].str[0]>60) & (df['location'].str[0]<=80),'zones']= 4
df.loc[ (df['location'].str[0]>80) & (df['location'].str[0]<=100),'zones']= 5
df.loc[ (df['location'].str[0]>100) & (df['location'].str[0]<=120),'zones']= 6
```

```python
def speedo_minutes_played(df0,player):
    
    a = np.zeros((df0.shape[0],1))
    num = df0[['match_id','type.name','tactics.lineup','player.name','card.name','substitution.replacement.name','minute']].to_numpy()
    num = np.append(num,a,axis=1)
            
    titularity=0
    been_changed=0
    been_entred=0
    been_changed_time=''
    been_excluded=''
    been_entred_time=''
    time_played='error'
    
    for i in range(0, len(num)):
        
        if num[i][1]=='Starting XI' and (player in str(num[0][2]) or player in str(num[1][2]) ) :
            titularity=1
        
        if num[i][1]=='Substitution' and num[i][3][0]==player:
            been_changed=1
            been_changed_time=num[i][6]
        
        if num[i][1]=='Foul Committed' and num[i][4]=='Red Card' and num[i][3]==player:
            been_excluded=1
            been_excluded_time=num[i][6]
    
        if num[i][1]=='Substitution'and num[i][5]==player:
            been_entred=1
            been_entred_time= num[i][6]
            
    if titularity==1:
        if been_changed==0:
            time_played=max(df0['minute'])
        if been_changed==1:
            if been_entred==0:
                time_played=been_changed_time
            if been_entred==1:
                time_played=been_changed_time+max(df0['minute'])-been_entred_time
        if been_excluded==1:
            time_played=been_excluded_time

    if titularity==0 and been_entred==1:
        if been_changed==0:
            time_played=max(df0['minute'])-been_entred_time
        if been_changed==1:
            time_played=been_changed_time-been_entred_time
        if been_excluded==1: 
            time_played=been_excluded_time-been_entred_time

    if titularity==0 and been_entred==0:
        time_played=0
          
    return df0['match_id'].unique()[0],int(df0.loc[df0['player.name'] == player,'player.id'].unique()[0]),player,time_played 
```

```python
%%time
speed_players_time = []
for group,subset in df.groupby(by='match_id'):
    #print(subset)
    for j in subset['player.name'].dropna().unique():
        #print("Match : ",group)        
        speed_players_time.append(speedo_minutes_played(subset,j))
```
