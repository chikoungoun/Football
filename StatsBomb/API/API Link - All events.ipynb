# Request the events

```python
all_events = []

for i in matches_list[162:167]:
    print(i)
    botola_events = "https://data.statsbombservices.com/api/v6/events/"+str(i)
    print(botola_events)
    response_events = requests.get(botola_events,auth=(username, password))
    print(response_events)
    events = response_events.json()
    df = pd.json_normalize(events)
    all_events.append(df)
print('\n ** DONE **')
```

```python
#concatenate all the dfs
pd.concat(all_events)
```
