# Match Score on Pitch

```python
import matplotlib.pyplot as plt
import seaborn as sns
from mplsoccer import Pitch
```

```python
team1 = "liverpool"
team2 = "Manchester city"
```

```python
import matplotlib.patheffects as path_effects
path_eff = [path_effects.Stroke(linewidth=3, foreground='black'),
            path_effects.Normal()]
```

```python
pitch = Pitch(pitch_type='statsbomb',pitch_color='#A2A0A5',line_color='black',stripe=False, line_zorder=2)

fig, ax = pitch.draw(figsize=(28,16))


# team names on the pitch
ax.text(30-(len(team1)/2),10,team1,fontsize=60,color='white')
ax.text(90-(len(team2)),10,team2,fontsize=60,color='white')

# team scores on the pitch
ax.text(30-(len(team1)/2),45,2,fontsize=190,color='white')
ax.text(90-(len(team2)/2),45,1,fontsize=190,color='white')

# team Xg on the pitch
ax.text(30-(len(team1)/2),70,1.00,fontsize=60,color='white')
ax.text(90-(len(team2)/2),70,2.38,fontsize=60,color='white')
```
