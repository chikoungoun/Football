# Match Score on Pitch

```python
import matplotlib.pyplot as plt
import seaborn as sns
from mplsoccer import Pitch
```

```python
team1 = "liverpool"
team2 = "Manchester city"
```

```python
pitch = Pitch(pitch_type='statsbomb',pitch_color='white',line_color='black',stripe=False, line_zorder=2)

fig, ax = pitch.draw(figsize=(28,16))


# team names on the pitch
home_name = ax.annotate(team1,(30-(len(team1)/2),10),fontsize=50)
visitor_name = ax.annotate(team2,(90-(len(team2)/2),10),fontsize=50)

# team scores on the pitch
home_score = ax.annotate(2,(30-(len(team1)/2),40),fontsize=190)
visitor_score = ax.annotate(1,(90-(len(team2)/2),40),fontsize=190)

# team Xg on the pitch
home_xg = ax.annotate(1.00,(30-(len(team1)/2),70),fontsize=90)
visitor_xg = ax.annotate(1.38,(90-(len(team2)/2),70),fontsize=90)

```
