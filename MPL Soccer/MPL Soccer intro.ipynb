### Plotting a blank pitch

```python
from mplsoccer.pitch import Pitch
pitch = Pitch(pitch_color='grass',line_color='white',stripe=True)
fig, ax = pitch.draw(figsize=(8,4))
```

### Use with matplotlib spec

```python
import matplotlib.pyplot as plt
```

draw as a sublplot 

```python
fig, axs = plt.subplots(nrows=1, ncols=2)
pitch = Pitch(pitch_color='red')
pie = axs[0].pie(x=[5, 15])
pitch.draw(ax=axs[1])
```

### Supported data providers + axis labels

mplsoccer supports 9 pitch types by specifying the pitch_type argument: ‘statsbomb’, ‘opta’, ‘tracab’, ‘wyscout’, ‘uefa’, ‘metricasports’, ‘custom’, ‘skillcorner’ and ‘secondspectrum’.

**Statsbomb axis**

```python
pitch = Pitch(pitch_type='statsbomb',pitch_length=105,pitch_width=68,
             axis=True,
             label=True) 
fig, ax = pitch.draw()
```

**tracab axis**

```python
pitch = Pitch(pitch_type='tracab',pitch_length=105,pitch_width=68,
             axis=True,
             label=True) 
fig, ax = pitch.draw()
```

**opta axis**

```python
pitch = Pitch(pitch_type='opta',pitch_length=105,pitch_width=68,
             axis=True,
             label=True) 
fig, ax = pitch.draw()
```
